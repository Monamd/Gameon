```python
def position_choice():
    choice='wrong'
    while choice not in ['0','1','2']:
        
        choice=input('Pick a postion (0,1,2): ')
        
        if choice not in ['0','1','2']:
            print('Sorry, invalid choice!')
        
    return int(choice)

    
    
```

```python
position_choice()
```

```python
def replacement_choice(game_list,postion):
    user_placement = input('type a string to place at postion: ')
    game_list[postion]=user_placement
    return game_list
```

```python
replacement_choice(game_list,1)
```

```python
def gameon_choice():
    
    choice='wrong'
    
    while choice not in ['Y','N']:
        
        choice=input('Keep playing? (Y or N) ')
        
        if choice not in ['Y','N']:
            print('Sorry, I dont understand, please choose Y or N!')
        
    if choice == 'Y':
        return True
    else:
        return False
       
```

```python
gameon_choice()
```

```python
game_on = True
game_list = [0,1,2]

while game_on:
    display_game(game_list)
    postion=position_choice()
    game_list=replacement_choice(game_list,postion)
    display_game(game_list)
    game_on=gameon_choice()
    
```
